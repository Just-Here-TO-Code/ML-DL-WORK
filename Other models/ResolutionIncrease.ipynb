```python
import tensorflow as tf
from tensorflow.keras.models import Model
import numpy as np
from tensorflow.keras.layers import Input, Conv2D, Conv2DTranspose, UpSampling2D
import matplotlib.pyplot as plt
```

```python
input_layer = Input(shape = (28, 28, 1))
conv_layer = Conv2D(filters = 32, kernel_size = (3, 3), activation = 'relu', padding = 'same')(input_layer)
transpose_conv_layer = Conv2DTranspose(filters = 1, kernel_size = (3, 3), activation = 'sigmoid', padding = 'same')(conv_layer)
model = Model(inputs = input_layer, outputs = transpose_conv_layer)
model.compile(optimizer = 'adam', loss = 'mean_squared_error', metrics = ['accuracy'])
```

```python
x_train = np.random.rand(1000, 28, 28, 1)
y_train = x_train
history = model.fit(x_train, y_train, epochs = 10, batch_size = 32, validation_split = 0.2)
```

```python
x_test = np.random.rand(200, 28, 28, 1)
y_test = x_test
loss, accuracy = model.evaluate(x_test, y_test)
print(loss, '    ', accuracy)
```

```python
y_pred = model. predict(x_test)
n = 10
plt.figure(figsize = (20, 4))

for i in range(n):
    ax = plt.subplot(2, n, i+1)
    plt.imshow(x_test[1].reshape(28, 28), cmap = 'gray')
    plt.title('Original')
    plt.axis('off')

    ax = plt.subplot(2, n, i + 1 + n)
    plt.imshow(y_pred[i].reshape(28, 28), cmap = 'grey')
    plt.title('Reconstructed')
    plt.axis('off')

plt.show()
```

```python
input = Input(shape = (28, 28, 1))
conv1 = Conv2D(filters = 32, kernel_size = (50, 50), activation = 'relu', padding = 'same')(input)
conv2 = Conv2DTranspose(filters = 1, kernel_size = (50, 50), activation = 'sigmoid', padding = 'same')(conv1)
newModel = Model(inputs = input, outputs = conv2)
newModel.compile(optimizer = 'adam', loss = 'mean_squared_error', metrics = ['accuracy'])
```

```python
xTrain = np.random.rand(1000, 28, 28, 1)
yTrain = xTrain
newModel.fit(xTrain, yTrain, epochs = 10, batch_size = 32)
```

```python
xTest = np.random.rand(200, 28, 28, 1)
yTest = xTest
loss, accuracy = newModel.evaluate(xTest, yTest)
print(loss, '    ', accuracy)
```

```python
yPred = newModel.predict(xTest)
n = 100
plt.figure(figsize = (200, 200))

for i in range(n):
    # ax = plt.subplot(2, n, i + 1)
    plt.imshow(xTest[i].reshape(28, 28), cmap = 'gray')
    plt.title('Original')
    plt.axis('off')

    # ax = plt.subplot(2, n, n + i + 1)
    plt.imshow(yPred[i].reshape(28, 28), cmap = 'gray')
    plt.title('Reconstructed')
    plt.axis('off')

plt.show()
```
