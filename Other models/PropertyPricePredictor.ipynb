```python
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
df = pd.read_csv('real_estate_data.csv')
```

```python
df.isna().sum()
df[0:10]
```

```python
df.dropna(inplace = True)
df[0:10]
```

```python
x = df.drop(columns = 'MEDV')
y = df['MEDV']
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = 0.2)
regTree = DecisionTreeRegressor(criterion = 'squared_error')
regTree.fit(xtrain, ytrain)
regTree.score(xtest, ytest)
```

```python
from sklearn import metrics
import numpy as np
predict = regTree.predict(xtest)
print(metrics.r2_score(ytest, predict))
```

```python
from sklearn.tree import plot_tree
from sklearn import tree
import matplotlib.pyplot as plt
tree.plot_tree(regTree)
plt.savefig('regTree.png', dpi = 1000)
plt.show()
```
